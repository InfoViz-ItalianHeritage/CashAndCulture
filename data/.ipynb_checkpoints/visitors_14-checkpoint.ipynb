```python
import pandas as pd

file = 'musei_tavola5_2014_visitatori.csv'
year="2014"

with open(file, "r", encoding = "utf-8") as f:
    for i,line in enumerate(f):
        if line.startswith("Regione"):
            header_row = i
            break
vis14 = pd.read_csv(file, skiprows = header_row, sep=";")
vis14
```

```python
vis14.rename(columns={"Totale complessivo":"Totale_regione"}, inplace= True)
vis14.dropna(axis=1, how="all", inplace= True)
vis14.dropna(axis=0, how="all", inplace= True)
# set 'Regione' as columns, months as rows
vis14 = vis14.set_index("Regione").T
vis14
```

```python
# rename the rows to add the year
vis14.index = [f"{month}_{year}" for month in vis14.index]
vis14
```
